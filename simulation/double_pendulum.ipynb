```python
import datetime

target_path = (
    f"../data/raw/double_pendulum/{datetime.datetime.now().strftime('%Y%m%d%H%M%S%f')}.csv"
)
```

```python
from recorder import Recorder

duration = 10  # (seconds)
framerate = 30  # (Hz)

rec = Recorder(model)

from math import ceil

n_steps = int(ceil(duration * framerate)) + 1

# Simulate and display video.
frames = []
mujoco.mj_resetData(model, data)  # Reset state and time.
import numpy as np

while data.time < duration:
    data.ctrl[0] = np.sin(data.time * 2) * 30
    data.ctrl[1] = np.sin(data.time * 10) * 30
    mujoco.mj_step(model, data)

    # record data
    rec.record(data)

    if len(frames) < data.time * framerate:
        renderer.update_scene(data)
        pixels = renderer.render()
        frames.append(pixels)

rec.save_df(target_path)

mediapy.show_video(frames, fps=framerate)
```
